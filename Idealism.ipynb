# Idealism

Is the phylosohpycal theory that, in the epistemic sense, claims that truth is independent of the minds of humans and other subjective observers.

To understand this theory we shall begin with a question: is there a universe which produces the same observable reality that we perceive, but does not contain the properties we believe it has?

To answer it we introduce our experiment. In which we shall attempt to produce one such "universe".

````python
%matplotlib widget

import sys

import numpy as np

from mpl_toolkits import mplot3d
from matplotlib import pyplot as plt
from PIL import Image

IM1 = 'logo-ro-web.jpg.webp'
IM2 = 'print-182373828.jpg'
IM3 = 'Costin-Raiciu-e1552558481123.jpg'

SIZE = 200
FILL = 1/10
````

````python
im1 = Image.open(IM1)
im2 = Image.open(IM2)
im3 = Image.open(IM3)
````

````python
im1 = im1.resize((SIZE, SIZE)).convert('L')
im2 = im2.resize((SIZE, SIZE)).convert('L')
im3 = im3.resize((SIZE, SIZE)).convert('L')

imo1 = np.asarray(im1, dtype=np.uint8) < 200
imo2 = np.asarray(im2, dtype=np.uint8) < 200
imo3 = np.asarray(im3, dtype=np.uint8) < 100
````

````python
im1 = np.stack([imo1 for _ in np.arange(SIZE)])
im2 = np.stack([imo2 for _ in np.arange(SIZE)])
im3 = np.stack([imo3 for _ in np.arange(SIZE)])

im1 = np.rot90(im1, axes=(0, 1))
im2 = np.rot90(im2, axes=(1, 2))
im3 = np.rot90(im3, axes=(0, 2))

u = np.ones(shape=(SIZE, SIZE, SIZE))

u = np.logical_and(u, ~im1)
u = np.logical_and(u, ~im2)
u = np.logical_and(u, ~im3)

u = (u * np.random.rand(SIZE, SIZE, SIZE)) >= (1 - FILL)
````

````python
u = np.argwhere(u)

x = u[:, 0]
y = u[:, 1]
z = u[:, 2]

fig = plt.figure()
ax = fig.add_subplot(projection='3d')

ax.axis('off')
ax.grid(visible=None)
ax.scatter3D(x, y, z, s=1)
ax.set_box_aspect([ub - lb for lb, ub in (getattr(ax, f'get_{a}lim')() for a in 'xyz')])
ax.view_init(0, 0)
ax.dist = 9

plt.show()
````

The above cube will serve as our universe. Intuitively we may expect that it has a dove carved into it from one side to another and the rest is full.

However, when rotated, we notice that there are other images carved into it. Not only that, but only one 10th of it is full.

Now we may ask a further question: How many universes are there that correspond to our perception, but do not correspond to our beliefs? To answer that question I may point the reader to the above code snippet:
```python
u = (u * np.random.rand(SIZE, SIZE, SIZE)) >= (1 - FILL))
```
Our universe has been randomly generated from a uniform distribution. And given that the images that are carved into the rest of the sides have been chose arbitrarily(almost), our intution has been contradictied not just randomly but in a way that has meaning to us.

Premises:
 - Our perception corresponds to multiple possible universes.
 - We fill in what's behind perception and produce an imaginary universe.

Thesis: Our claims regarding the imaginary universe may not hold in unexpected ways, starkly contradicting our intuition.

Experiment:
  - Create a visual universe consisting of a cube that when projected unto a surface produces an intuition of what the rest of the universe may look like.
  - Contradict the intuition in an unxpected way.
  - Demonstrate that there are many such contradictions possible.
  
How does the current draft demonstrate this?
  - The cube produces the intuition that it is full and that it has sculpted into it a certain image.
  - It contradicts intuition at closer inspection when it is revealed that it is mostly hollow and that other images are sculpted into it as well.
  - It demonstrates that many such contradictions are possible by 1) the posibility of having sculpted into it arbitrary images and 2) by the uniform distribution of its filling.
